### Task 03. Provide the Instacart senior stakeholders with descriptive findings about sales. Create a histogram of the order_hour_of_day column

```python
#Checking the histogram accuracy
instacart['order_hour_of_day'].value_counts(dropna=False).sort_index()
```

#### Distribution of Orders by Hour of the Day

#### The histogram above illustrates the distribution of Instacart orders across different hours of the day. Here are the key observations:
#### - There is a clear peak in order activity during the mid-morning to early afternoon hours (9 AM to 3 PM), with the highest number of orders occurring at 10 AM.
#### - Order activity starts to increase significantly from 7 AM, reaching its peak between 9 AM and 3 PM.
#### - After 3 PM, the number of orders gradually decreases, with a notable decline from 5 PM onwards.
#### - The lowest order volumes are observed in the early morning hours (midnight to 6 AM), with the least activity around 3 AM.
#### This distribution suggests that Instacart's customers are most active during typical waking hours, particularly around mid-morning to early afternoon. This information can be valuable for optimizing staffing and operational planning to handle peak times efficiently.

### Task 06. Determine whether theres a connection between age and family situation by creating a line chart exploring the connections between age and number of dependents

#####  Connection Between Age and Number of Dependents

#### The line chart above explores the connection between age and the number of dependents for a small sample of customers. Here are the key observations:
#### - At age 20, the number of dependents is 3.
#### - As the age increases to 40, the number of dependents remains at 3.
#### - At age 61, the number of dependents decreases to 1.
#### - Similarly, at age 66, the number of dependents is 2.
#### - At age 81, the number of dependents decreases back to 1. 
#### This line chart suggests that the number of dependents tends to vary with age, potentially peaking during the mid-life years and decreasing as age increases further. This information can be valuable for targeted marketing campaigns, as it provides insights into the family situations of different age groups.


### Task 07. Explore whether theres a connection between age and spending power (income) using a scatterplot

```python
instacart_4.head(10)
```

 #### There is a wide range of incomes among the younger age groups, particularly from ages 20 to around 40, with a concentration at the lower end of the income scale. 
 #### Starting from around age 40 and onwards, there appears to be an upward trend in income, with many of the data points concentrated in the higher income brackets. 
 #### This could suggest that income increases with age. The densest concentration of points for individuals over 40 is significantly higher on the income scale 
 #### compared to those under 40, which could indicate that the median income increases with age. The scatterplot provides a visual representation that can hint 
 #### at certain trends, like potential increases in income with age, but it would require statistical analysis to confirm any correlations or to understand 
 #### This scatterplot suggests that there is no clear linear relationship between age and income in this sample. However, it does indicate that individuals in their mid to late 60s might have higher earning potential compared to younger and older age groups. These insights can be useful for tailoring marketing strategies based on the spending power of different age demographics.
